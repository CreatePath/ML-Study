# 활성화 함수 구현

## 1. 계단 함수

```python
# 계단 함수 (step function) 구현
def step_function(x):
#     if x > 0:
#         return 1
#     else: 
#         return 0
    y = x > 0 # np.array 도 입력받을 수 있다!
    return y.astype(np.int64)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 2. 시그모이드 함수

```python
# 시그모이드 함수 (sigmoid function) 구현
def sigmoid(x): # x가 np.array 여도 상관 없음.
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.show()
```

## 3. 렐루 함수

```python
# 렐루 함수 (ReLU. Rectified linear Unit) 구현
def relu(x):
    return np.maximum(0, x)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x,y)
plt.ylim(-0.1, 5.0)
plt.show()
```
